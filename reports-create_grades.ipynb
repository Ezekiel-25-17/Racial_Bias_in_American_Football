```python
import pandas as pd
import os
```

```python
path = 'dataset-quarterback_scouting_reports'
filename = 'qb_draft_reports_tagged.csv'

df = pd.read_csv(os.path.join(path, filename))
df.head(3)
```

```python
df0 = df[(df['InjuryConcerns'] == 0) & (df['CharacterIssues'] == 0) & (df['Armstrength_Negative'] == 0) 
       & (df['Armstrength_Positive'] == 0) & (df['PlayMaker_Negative'] == 0) & (df['PlayMaker_Positive'] == 0) 
       & (df['Rare_Gifted_Negative'] == 0) & (df['Rare_Gifted_Positive'] == 0) & (df['Improvisation_Negative'] == 0)
       & (df['Improvisation_Positive'] == 0) & (df['Release_Negative'] == 0) & (df['Release_Positive'] == 0)
       & (df['Mechanics_Negative'] == 0) & (df['Mechanics_Positive'] == 0) & (df['Footwork_Negative'] == 0)
       & (df['Footwork_Positive'] == 0) & (df['Accuracy_Negative'] == 0) & (df['Accuracy_Positive'] == 0)
       & (df['PocketPresence_Negative'] == 0) & (df['PocketPresence_Positive'] == 0)
       & (df['FootballIQ_Negative'] == 0) & (df['FootballIQ_Positive'] == 0) & (df['Leadership_Negative'] == 0)
       & (df['Leadership_Positive'] == 0) & (df['Competitor_Negative'] == 0) & (df['Competitor_Positive'] == 0)
       & (df['Toughness_Negative'] == 0) & (df['Toughness_Positive'] == 0) & (df['Clutch_Negative'] == 0)
       & (df['Clutch_Positive'] == 0) & (df['Confidence_Negative'] == 0) & (df['Confidence_Positive'] == 0)
       & (df['Decision_Making_Negative'] == 0) & (df['Decision_Making_Positive'] == 0)
       & (df['ReadingDefense_Negative'] == 0) & (df['ReadingDefense_Positive'] == 0)
       & (df['Progressions_Negative'] == 0) & (df['Progressions_Positive'] == 0)
       & (df['RunningAbility_Negative'] == 0) & (df['RunningAbility_Positive'] == 0)
       & (df['Athletic_Negative'] == 0) & (df['Athletic_Positive'] == 0) & (df['Escapability_Negative'] == 0)
       & (df['Escapability_Positive'] == 0)]

df0.head(3)
```

```python
df = df[~df.index.isin(df0.index)].reset_index(drop=True)
df.head(3)
```

```python
df = df.groupby(['Player', 'Year']).sum(numeric_only=True).reset_index().sort_values('Year').reset_index(drop=True)
df.head(3)
```

```python
df.insert(2, 'Total_Reports', 0)

df['Total_Reports'] = df.loc[:, 'InjuryConcerns':].sum(axis=1)
df.head(3)
```

```python
df = df[df['Total_Reports'] >= 10].reset_index(drop=True)
df.head(3)
```

```python
path = 'dataset-quarterback_stats'
filename = 'census.csv'

# Import info about players ethnicity
df2 = pd.read_csv(os.path.join(path, filename))

# Merge df and df2
df = df.merge(df2, how='left', on='Player')

df.head(3)
```

```python
df = df[['Player', 'Year', 'Race', 'Total_Reports', 'InjuryConcerns', 'CharacterIssues',
       'Armstrength_Negative', 'Armstrength_Positive', 'PlayMaker_Negative',
       'PlayMaker_Positive', 'Rare_Gifted_Negative', 'Rare_Gifted_Positive',
       'Improvisation_Negative', 'Improvisation_Positive', 'Release_Negative',
       'Release_Positive', 'Mechanics_Negative', 'Mechanics_Positive',
       'Footwork_Negative', 'Footwork_Positive', 'Accuracy_Negative',
       'Accuracy_Positive', 'PocketPresence_Negative',
       'PocketPresence_Positive', 'FootballIQ_Negative', 'FootballIQ_Positive',
       'Leadership_Negative', 'Leadership_Positive', 'Competitor_Negative',
       'Competitor_Positive', 'Toughness_Negative', 'Toughness_Positive',
       'Clutch_Negative', 'Clutch_Positive', 'Confidence_Negative',
       'Confidence_Positive', 'Decision_Making_Negative',
       'Decision_Making_Positive', 'ReadingDefense_Negative',
       'ReadingDefense_Positive', 'Progressions_Negative',
       'Progressions_Positive', 'RunningAbility_Negative',
       'RunningAbility_Positive', 'Athletic_Negative', 'Athletic_Positive',
       'Escapability_Negative', 'Escapability_Positive']].rename(columns={'Total_Reports':'Report_Count'})

df.head(5)
```

```python
d = {'Player' : df['Player'], 
    'Year': df['Year'], 
    'Race' : df['Race'], 
    'Report_Count' : df['Report_Count'],

    'Mental_Count' : df['Leadership_Negative'] + df['Leadership_Positive'] + df['Competitor_Negative'] + df['Competitor_Positive'] 
                   + df['Toughness_Negative'] + df['Toughness_Positive'] + df['Clutch_Negative'] + df['Clutch_Positive'] 
                   + df['Confidence_Negative'] + df['Confidence_Positive'] + df['CharacterIssues'],
    'Mental_Positive' : df['Leadership_Positive'] + df['Competitor_Positive'] + df['Toughness_Positive'] 
                      + df['Clutch_Positive'] + df['Confidence_Positive'],
    'Mental_Negative' : df['Leadership_Negative'] + df['Competitor_Negative'] + df['Toughness_Negative'] 
                      + df['Clutch_Negative'] + df['Confidence_Negative'] + df['CharacterIssues'],
    'Mental_Net' : - df['Leadership_Negative'] + df['Leadership_Positive'] - df['Competitor_Negative'] + df['Competitor_Positive'] 
                   - df['Toughness_Negative'] + df['Toughness_Positive'] - df['Clutch_Negative'] + df['Clutch_Positive'] 
                   - df['Confidence_Negative'] + df['Confidence_Positive'] - df['CharacterIssues'],

    'Athleticism_Count' : df['RunningAbility_Negative'] + df['RunningAbility_Positive'] + df['Athletic_Negative'] + df['Athletic_Positive']
                        + df['Escapability_Negative'] + df['Escapability_Positive'] + df['InjuryConcerns'],
    'Athleticism_Positive' : df['RunningAbility_Positive'] + df['Athletic_Positive'] + df['Escapability_Positive'],
    'Athleticism_Negative' : df['RunningAbility_Negative'] + df['Athletic_Negative'] + df['Escapability_Negative'] + df['InjuryConcerns'],
    'Athleticism_Net' : - df['RunningAbility_Negative'] + df['RunningAbility_Positive'] - df['Athletic_Negative'] + df['Athletic_Positive']
                        - df['Escapability_Negative'] + df['Escapability_Positive'] - df['InjuryConcerns'],
    
    'Tecnique_Count' : df['Release_Negative'] + df['Release_Positive'] + df['Mechanics_Negative'] + df['Mechanics_Positive']
                     + df['Footwork_Negative'] + df['Footwork_Positive'] + df['Accuracy_Negative'] + df['Accuracy_Positive'],
    'Tecnique_Positive' : df['Release_Positive'] + df['Mechanics_Positive'] + df['Footwork_Positive'] + df['Accuracy_Positive'],
    'Tecnique_Negative' : df['Release_Negative'] + df['Mechanics_Negative'] + df['Footwork_Negative'] + df['Accuracy_Negative'],
    'Tecnique_Net' : - df['Release_Negative'] + df['Release_Positive'] - df['Mechanics_Negative'] + df['Mechanics_Positive']
                     - df['Footwork_Negative'] + df['Footwork_Positive'] - df['Accuracy_Negative'] + df['Accuracy_Positive'],
    
    'NaturalTalent_Count' : df['Armstrength_Negative'] + df['Armstrength_Positive'] + df['PlayMaker_Negative'] + df['PlayMaker_Positive'] 
	                      + df['Improvisation_Negative'] + df['Improvisation_Positive'] + df['Rare_Gifted_Negative'] + df['Rare_Gifted_Positive'], 
    'NaturalTalent_Positive' : df['Armstrength_Positive'] + df['PlayMaker_Positive'] + df['Improvisation_Positive'] + df['Rare_Gifted_Positive'], 
    'NaturalTalent_Negative' : df['Armstrength_Negative'] + df['PlayMaker_Negative'] + df['Improvisation_Negative'] + df['Rare_Gifted_Negative'], 
    'NaturalTalent_Net' : - df['Armstrength_Negative'] + df['Armstrength_Positive'] - df['PlayMaker_Negative'] + df['PlayMaker_Positive'] 
	                      - df['Improvisation_Negative'] + df['Improvisation_Positive'] - df['Rare_Gifted_Negative'] + df['Rare_Gifted_Positive'],

    'FootballIntelligence_Count' : df['FootballIQ_Negative'] + df['FootballIQ_Positive'] + df['Decision_Making_Negative'] + df['Decision_Making_Positive'] 
                                 + df['ReadingDefense_Negative'] + df['ReadingDefense_Positive'] + df['Progressions_Negative'] + df['Progressions_Positive']
	                             + df['PocketPresence_Negative'] + df['PocketPresence_Positive'],
    'FootballIntelligence_Positive' : df['FootballIQ_Positive'] + df['Decision_Making_Positive'] + df['ReadingDefense_Positive'] 
                                    + df['Progressions_Positive'] + df['PocketPresence_Positive'],
    'FootballIntelligence_Negative' : + df['FootballIQ_Negative'] + df['Decision_Making_Negative'] + df['ReadingDefense_Negative']
                                      + df['Progressions_Negative'] + df['PocketPresence_Negative'],
    'FootballIntelligence_Net' : - df['FootballIQ_Negative'] + df['FootballIQ_Positive'] - df['Decision_Making_Negative'] + df['Decision_Making_Positive'] 
                                 - df['ReadingDefense_Negative'] + df['ReadingDefense_Positive'] - df['Progressions_Negative'] + df['Progressions_Positive']
	                             - df['PocketPresence_Negative'] + df['PocketPresence_Positive']}

newdf = pd.DataFrame(d)
newdf
```

```python
# Dividing by Count

newdf['Mental_Positive'] = newdf['Mental_Positive'] / newdf['Mental_Count']
newdf['Mental_Negative'] = newdf['Mental_Negative'] / newdf['Mental_Count']
newdf['Mental_Net'] = newdf['Mental_Net'] / newdf['Mental_Count']

newdf['Athleticism_Positive'] = newdf['Athleticism_Positive'] / newdf['Athleticism_Count']
newdf['Athleticism_Negative'] = newdf['Athleticism_Negative'] / newdf['Athleticism_Count']
newdf['Athleticism_Net'] = newdf['Athleticism_Net'] / newdf['Athleticism_Count']

newdf['Tecnique_Positive'] = newdf['Tecnique_Positive'] / newdf['Tecnique_Count']
newdf['Tecnique_Negative'] = newdf['Tecnique_Negative'] / newdf['Tecnique_Count']
newdf['Tecnique_Net'] = newdf['Tecnique_Net'] / newdf['Tecnique_Count']

newdf['NaturalTalent_Positive'] = newdf['NaturalTalent_Positive'] / newdf['NaturalTalent_Count']
newdf['NaturalTalent_Negative'] = newdf['NaturalTalent_Negative'] / newdf['NaturalTalent_Count']
newdf['NaturalTalent_Net'] = newdf['NaturalTalent_Net'] / newdf['NaturalTalent_Count']

newdf['FootballIntelligence_Positive'] = newdf['FootballIntelligence_Positive'] / newdf['FootballIntelligence_Count']
newdf['FootballIntelligence_Negative'] = newdf['FootballIntelligence_Negative'] / newdf['FootballIntelligence_Count']
newdf['FootballIntelligence_Net'] = newdf['FootballIntelligence_Net'] / newdf['FootballIntelligence_Count']

newdf = newdf.fillna(0)
newdf
```

```python
path = 'dataset-quarterback_scouting_reports'
filename = 'qb_grading.csv'

newdf.to_csv(os.path.join(path, filename), index=False)
```
