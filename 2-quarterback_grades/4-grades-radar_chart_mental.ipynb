```python
path = 'dataset-grades'
filename = 'qb_grades_mental.csv'

df = pd.read_csv(os.path.join(path, filename))
df.head(3)
```

```python
# Separate White and Non-White Players
w = df[df['Race'] == 'White']
nw = df[df['Race'] == 'Non-White']
```

```python
d = {'Title' : ['Average QB', 'Average White QB', 'Average Non-White QB'],
     'Leadership_Net' : [df['Leadership_Net'].mean(), w['Leadership_Net'].mean(), nw['Leadership_Net'].mean()],
     'Competitor_Net' : [df['Competitor_Net'].mean(), w['Competitor_Net'].mean(), nw['Competitor_Net'].mean()],
     'Toughness_Net' : [df['Toughness_Net'].mean(), w['Toughness_Net'].mean(), nw['Toughness_Net'].mean()],
     'Clutch_Net' : [df['Clutch_Net'].mean(), w['Clutch_Net'].mean(), nw['Clutch_Net'].mean()],
     'Confidence_Net' : [df['Confidence_Net'].mean(), w['Confidence_Net'].mean(), nw['Confidence_Net'].mean()]}

df_net = pd.DataFrame(d)
df_net = df_net[['Title', 'Leadership_Net', 'Competitor_Net', 'Toughness_Net', 'Clutch_Net', 'Confidence_Net']]
df_net
```

```python
d = {'Title' : ['Average QB', 'Average White QB', 'Average Non-White QB'],
     'Leadership_Positive' : [df['Leadership_Positive'].mean(), w['Leadership_Positive'].mean(), nw['Leadership_Positive'].mean()],
     'Competitor_Positive' : [df['Competitor_Positive'].mean(), w['Competitor_Positive'].mean(), nw['Competitor_Positive'].mean()],
     'Toughness_Positive' : [df['Toughness_Positive'].mean(), w['Toughness_Positive'].mean(), nw['Toughness_Positive'].mean()],
     'Clutch_Positive' : [df['Clutch_Positive'].mean(), w['Clutch_Positive'].mean(), nw['Clutch_Positive'].mean()],
     'Confidence_Positive' : [df['Confidence_Positive'].mean(), w['Confidence_Positive'].mean(), nw['Confidence_Positive'].mean()]}

df_pos = pd.DataFrame(d)
df_pos = df_pos[['Title', 'Leadership_Positive', 'Competitor_Positive', 'Toughness_Positive', 'Clutch_Positive', 'Confidence_Positive']]
df_pos
```

```python
d = {'Title' : ['Average QB', 'Average White QB', 'Average Non-White QB'],
     'Leadership_Negative' : [df['Leadership_Negative'].mean(), w['Leadership_Negative'].mean(), nw['Leadership_Negative'].mean()],
     'Competitor_Negative' : [df['Competitor_Negative'].mean(), w['Competitor_Negative'].mean(), nw['Competitor_Negative'].mean()],
     'Toughness_Negative' : [df['Toughness_Negative'].mean(), w['Toughness_Negative'].mean(), nw['Toughness_Negative'].mean()],
     'Clutch_Negative' : [df['Clutch_Negative'].mean(), w['Clutch_Negative'].mean(), nw['Clutch_Negative'].mean()],
     'Confidence_Negative' : [df['Confidence_Negative'].mean(), w['Confidence_Negative'].mean(), nw['Confidence_Negative'].mean()]}

df_neg = pd.DataFrame(d)
df_neg = df_neg[['Title', 'Leadership_Negative', 'Competitor_Negative', 'Toughness_Negative', 'Clutch_Negative', 'Confidence_Negative']]
df_neg
```

```python
categories = ['Leadership', 'Competitor', 'Toughness', 'Clutch', 'Confidence']
```

```python
# Calculate angles for radar chart
offset = np.pi/5
angles = np.linspace(0, 2*np.pi, len(categories) + 1) + offset
```

```python
# Function to create radar chart
def create_radar_chart(ax, angles, player_data, categories, color='#003594', title='', title_flag=False):
    
    # Plot data and fill with team color
    ax.plot(angles, np.append(player_data[-(len(angles)-1):], player_data[-(len(angles)-1)]), color=color, linewidth=2)
    ax.fill(angles, np.append(player_data[-(len(angles)-1):], player_data[-(len(angles)-1)]), color=color, alpha=0.2)
    
    # Set category labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=40)
    
    # Remove radial labels
    ax.set_yticklabels([])

    # Add Title
    ax.text(np.pi/2, 1.4, player_data[0], ha='center', va='center', size=50, color=color)
    
    if title_flag:
        ax.text(np.pi/2, 1.70, title, ha='center', va='center', size=55)
    
    # Use white grid
    ax.grid(color='white', linewidth=1.5)

    # Set axis limits
    ax.set(xlim=(0, 2*np.pi), ylim=(0, 1))

    return ax

def get_row_data(data, row):
    return np.asarray(data[data['Title'] == row])[0]
```

```python
# General plot parameters
plt.rcParams['font.size'] = 16
plt.rcParams['axes.linewidth'] = 0
plt.rcParams['xtick.major.pad'] = 15
plt.rcParams['font.family'] = 'Avenir'

# Create figure
fig = plt.figure(figsize=(40, 40), facecolor='white')

# Add subplots
ax1 = fig.add_subplot(131, projection='polar', facecolor='#ededed')
ax2 = fig.add_subplot(132, projection='polar', facecolor='#ededed')
ax3 = fig.add_subplot(133, projection='polar', facecolor='#ededed')

# Adjust space between subplots
plt.subplots_adjust(hspace=0.8, wspace=0.5)

# Get QB data
all_data = get_row_data(df_net, 'Average QB')
w_data = get_row_data(df_net, 'Average White QB')
nw_data = get_row_data(df_net, 'Average Non-White QB')

# Plot QB data
ax1 = create_radar_chart(ax1, angles, all_data, categories)
ax2 = create_radar_chart(ax2, angles, w_data, categories, title='Net Grades: Mental', title_flag=True)
ax3 = create_radar_chart(ax3, angles, nw_data, categories)

plt.show()
```
