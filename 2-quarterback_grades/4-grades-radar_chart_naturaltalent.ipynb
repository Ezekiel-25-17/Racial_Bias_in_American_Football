```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os
```

```python
path = 'dataset-grades'
filename = 'qb_grades_naturaltalent.csv'

df = pd.read_csv(os.path.join(path, filename))
df.head(3)
```

```python
""" GRADING SYSTEM

'Natural_Talent'
       'Armstrength_Negative', 'Armstrength_Positive', 
       'PlayMaker_Negative', 'PlayMaker_Positive', 
	'Improvisation_Negative', 'Improvisation_Positive', 
       'Rare_Gifted_Negative', 'Rare_Gifted_Positive', 
"""
```

```python
# Separate White and Non-White Players
w = df[df['Race'] == 'White']
nw = df[df['Race'] == 'Non-White']
```

```python
d = {'Title' : ['Average QB', 'Average White QB', 'Average Non-White QB'],
     'Armstrength_Net' : [df['Armstrength_Net'].mean(), w['Armstrength_Net'].mean(), nw['Armstrength_Net'].mean()],
     'PlayMaker_Net' : [df['PlayMaker_Net'].mean(), w['PlayMaker_Net'].mean(), nw['PlayMaker_Net'].mean()],
     'Improvisation_Net' : [df['Improvisation_Net'].mean(), w['Improvisation_Net'].mean(), nw['Improvisation_Net'].mean()],
     'Rare_Gifted_Net' : [df['Rare_Gifted_Net'].mean(), w['Rare_Gifted_Net'].mean(), nw['Rare_Gifted_Net'].mean()]}

df_net = pd.DataFrame(d)
df_net = df_net[['Title', 'Armstrength_Net', 'PlayMaker_Net', 'Improvisation_Net', 'Rare_Gifted_Net']]
df_net
```

```python
d = {'Title' : ['Average QB', 'Average White QB', 'Average Non-White QB'],
     'Armstrength_Positive' : [df['Armstrength_Positive'].mean(), w['Armstrength_Positive'].mean(), nw['Armstrength_Positive'].mean()],
     'PlayMaker_Positive' : [df['PlayMaker_Positive'].mean(), w['PlayMaker_Positive'].mean(), nw['PlayMaker_Positive'].mean()],
     'Improvisation_Positive' : [df['Improvisation_Positive'].mean(), w['Improvisation_Positive'].mean(), nw['Improvisation_Positive'].mean()],
     'Rare_Gifted_Positive' : [df['Rare_Gifted_Positive'].mean(), w['Rare_Gifted_Positive'].mean(), nw['Rare_Gifted_Positive'].mean()]}

df_pos = pd.DataFrame(d)
df_pos = df_pos[['Title', 'Armstrength_Positive', 'PlayMaker_Positive', 'Improvisation_Positive', 'Rare_Gifted_Positive']]
df_pos
```

```python
d = {'Title' : ['Average QB', 'Average White QB', 'Average Non-White QB'],
     'Armstrength_Negative' : [df['Armstrength_Negative'].mean(), w['Armstrength_Negative'].mean(), nw['Armstrength_Negative'].mean()],
     'PlayMaker_Negative' : [df['PlayMaker_Negative'].mean(), w['PlayMaker_Negative'].mean(), nw['PlayMaker_Negative'].mean()],
     'Improvisation_Negative' : [df['Improvisation_Negative'].mean(), w['Improvisation_Negative'].mean(), nw['Improvisation_Negative'].mean()],
     'Rare_Gifted_Negative' : [df['Rare_Gifted_Negative'].mean(), w['Rare_Gifted_Negative'].mean(), nw['Rare_Gifted_Negative'].mean()]}

df_neg = pd.DataFrame(d)
df_neg = df_neg[['Title', 'Armstrength_Negative', 'PlayMaker_Negative', 'Improvisation_Negative', 'Rare_Gifted_Negative']]
df_neg
```

```python
categories = ['Armstrength', 'PlayMaker', 'Improvisation', 'Rare_Gifted']
```

```python
# Calculate angles for radar chart
offset = np.pi/4
angles = np.linspace(0, 2*np.pi, len(categories) + 1) + offset
```

```python
# Function to create radar chart
def create_radar_chart(ax, angles, player_data, categories, color='#003594', title='', title_flag=False):
    
    # Plot data and fill with team color
    ax.plot(angles, np.append(player_data[-(len(angles)-1):], player_data[-(len(angles)-1)]), color=color, linewidth=2)
    ax.fill(angles, np.append(player_data[-(len(angles)-1):], player_data[-(len(angles)-1)]), color=color, alpha=0.2)
    
    # Set category labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=40)
    
    # Remove radial labels
    ax.set_yticklabels([])

    # Add Title
    ax.text(np.pi/2, 1.4, player_data[0], ha='center', va='center', size=50, color=color)
    
    if title_flag:
        ax.text(np.pi/2, 1.70, title, ha='center', va='center', size=55)
    
    # Use white grid
    ax.grid(color='white', linewidth=1.5)

    # Set axis limits
    ax.set(xlim=(0, 2*np.pi), ylim=(0, 1))

    return ax

def get_row_data(data, row):
    return np.asarray(data[data['Title'] == row])[0]
```

```python
# General plot parameters
plt.rcParams['font.size'] = 16
plt.rcParams['axes.linewidth'] = 0
plt.rcParams['xtick.major.pad'] = 15
plt.rcParams['font.family'] = 'Avenir'

# Create figure
fig = plt.figure(figsize=(40, 40), facecolor='white')

# Add subplots
ax1 = fig.add_subplot(131, projection='polar', facecolor='#ededed')
ax2 = fig.add_subplot(132, projection='polar', facecolor='#ededed')
ax3 = fig.add_subplot(133, projection='polar', facecolor='#ededed')

# Adjust space between subplots
plt.subplots_adjust(hspace=0.8, wspace=0.5)

# Get QB data
all_data = get_row_data(df_net, 'Average QB')
w_data = get_row_data(df_net, 'Average White QB')
nw_data = get_row_data(df_net, 'Average Non-White QB')

# Plot QB data
ax1 = create_radar_chart(ax1, angles, all_data, categories)
ax2 = create_radar_chart(ax2, angles, w_data, categories, title='Net Grades: Natural Talent', title_flag=True)
ax3 = create_radar_chart(ax3, angles, nw_data, categories)

plt.show()
```
